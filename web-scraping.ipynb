# Practice scraping data from a real website

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.sebikes.com/products/2021-big-flyer-29"
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
rows = soup.find_all('tr')
```

```python
row_table = [title.text for title in rows]
print(row_table)
```
